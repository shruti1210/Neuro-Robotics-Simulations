```python
import nest
import nest.voltage_trace
#set verbosity to suppress info messages 
#reset the kernel to allow execute the script several times in Python shell 
#without interferences from previous NEST simulations
nest.set_verbosity("M_WARNING")
nest.ResetKernel()
```

```python
#Create = adds neurons and devices 
neuron = nest.Create("iaf_psc_alpha")
voltmeter = nest.Create("voltmeter")
```

```python
nest.SetStatus(neuron, "I_e", 376.0)
nest.SetStatus(voltmeter, [{"withgid": True}])
```

```python
nest.Connect(voltmeter, neuron)
```

```python
nest.Simulate(1000.0)
```

```python
nest.voltage_trace.from_device(voltmeter)
```
